```python
# import modules
import cv2
import mediapipe as mp
import time
```

```python
# get camera video
cap = cv2.VideoCapture(0)
```

```python
# mediapipe hands detection
mpHands = mp.solutions.hands
hands = mpHands.Hands(static_image_mode=False,max_num_hands = 2,min_detection_confidence = 0.5, min_tracking_confidence = 0.5)
mpDraw = mp.solutions.drawing_utils # to draw the hand
```

```python
# parameters
previous_time = 0
current_time = 0
```

```python
# process each frame

while True:
    sucess,frame = cap.read()

    # process image to find hand
    frameRGB = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    results = hands.process(frameRGB)
    print(results.multi_hand_landmarks) # print  if there is a hand or not

    (height,width,channels) = frame.shape

    # exctract parameters
    if results.multi_hand_landmarks:
        for handLms in results.multi_hand_landmarks:
            for id, lm in enumerate(handLms.landmark):
                # print(id,lm)
                cx,cy = int(lm.x*width),int(lm.y*height) # place in pixel of each points of the hand 
                print(f'landmark number {id} : column = {cx} row = {cy}')
                
                if id == 0 : # only looking for the first landmark
                    cv2.circle(frame,(cx,cy),25,(255,0,255),cv2.FILLED)
                    
            mpDraw.draw_landmarks(frame,handLms, mpHands.HAND_CONNECTIONS)

    current_time = time.time()
    fps = 1/(current_time-previous_time)
    previous_time=current_time

    cv2.putText(frame,str(int(fps)),(10,70),cv2.FONT_HERSHEY_PLAIN,3,(255,0,255),3)

    # display
    cv2.imshow('Video',frame)
    cv2.waitKey(1)
```
